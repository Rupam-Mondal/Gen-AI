```python
#Merge dataframe
```

```python
df1 = pd.DataFrame({
    'key1':[1,  2 , 3 , 4 , 5],
    'key2':[5 , 6 , 7 , 8 , 9],
    'key3':[9 , 8 , 1 , 3 , 4]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'key1':[1,  2 , 45 , 4 , 5],
    'key2':[56 , 6 , 7 , 8 , 9],
    'key3':[9 , 89 , 12 , 3 , 4]
})
```

```python
df2
```

```python
pd.merge(df1 , df2 , how='inner')
```

```python
pd.merge(df1 , df2 , how='left')
```

```python
pd.merge(df1 , df2 , how='outer')
```

```python
pd.merge(df1 , df2 , how='cross')
```

```python
pd.merge(df1 , df2 , how='left')
```

```python
pd.merge(df1 , df2 , how='left' , left_on='key2' , right_on='key1')
```
